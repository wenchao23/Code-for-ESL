```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('prostate.data', sep="\t") 
print(df.head())
print(list(df))
target = "lpsa"
feature = ['lcavol','lweight','age','lbph','svi','lcp',"gleason",'pgg45']
is_train = df["train"] == "T"
X, y = df[feature].values, df[target].values
X_train,y_train = X[is_train], y[is_train]
X_test,y_test = X[~is_train], y[~is_train]
N, p = X_train.shape

class Regression:
    def __initial__(self):
        pass
    
    def X_arg(self,X):
        N, _ = X.shape
        X = np.concatenate((np.ones((N,1)),X), axis=1)
        return X
    
    def scalar(self,X):
        X_mean = np.mean(X,axis=0)
        X_std = np.std(X,axis=0)
        X = (X-X_mean)/X_std
        return X_mean, X_std
    
    def OLS(self, X, Y, *arg):
        beta_hat = np.linalg.inv(X.transpose()@X)@X.transpose()@(Y-np.mean(Y))       
        beta_hat = np.insert(beta_hat,0,np.mean(Y))
        return beta_hat

    def Ridge(self, X, Y, Lambda): # Ridge regression
        _, p = X.shape
        I = np.eye(p)
        beta_hat = np.linalg.inv(X.transpose()@X+Lambda*I)@X.transpose()@(Y-np.mean(Y))   
        beta_hat = np.insert(beta_hat,0,np.mean(y_train))

        return beta_hat
    
   
    def Lasso(self,X,Y,Lambda,maxit=1000): # Lasso regression using coordinate descent algorithms
        
        theta = self.Ridge(X, Y, Lambda)
        theta = theta[1:]
        _, p = X.shape
        for i in np.arange(0,maxit):
            a = np.zeros(p)
            c = np.zeros(p)

            for j in np.arange(0,p):
                a[j] = 2*sum(X[:,j]**2)
                for k in np.arange(0,p):
                    wx = theta[j]*X[:,j]
                    c[j] = 2*sum(((Y-np.mean(Y))-X@theta+wx)*X[:,j])
                    theta[j] = self.soft(c[j]/a[j],Lambda/a[j])
        theta = np.insert(theta,0,np.mean(Y))                
        return theta
    
    
    def soft(self,a,b):
        if a< -b:
            theta = a+b
        elif a>=-b and a<=b:
            theta = 0
        elif a>b:
            theta = a-b
        return theta
    
    def IFS(self,X,Y, iter = 100): #Incremental FS
        _, p = X.shape
        r = Y - np.mean(Y)
        beta = np.zeros(p)
        path = np.empty((0,p))
        ep = 0.01
        for i in np.arange(0,iter):
            j = r.transpose()@X
            jk = np.argmax(abs(j))
            beta[jk] = beta[jk] + ep*np.sign(j[jk])
            r = r - ep*np.sign(j[jk])*X[:,jk]
            path = np.vstack((path,beta))
            
        return path
        
    def CV(self, model, splits,X,Y,*arg): # Cross Validation
        
        n_splits, _ = splits.shape

        RSS = np.array([])
        
        for i in np.arange(0,n_splits):
            val_index = splits[i,:]
            X_val = X[val_index,:]
            Y_val = Y[val_index]
            X_train = np.delete(X, val_index, 0)
            Y_train = np.delete(Y, val_index, 0)
            beta = model(X_train,Y_train,*arg)
            pred = self.X_arg(X_val)@beta

            RSS1 = np.sqrt((np.mean((Y_val-pred)**2)))

            RSS = np.append(RSS,RSS1)
        return RSS
    def split(self, X, n_folds = 10, seed = 10): #Shuffle split
        N, _ = X.shape
        np.random.seed(seed)
        

        number_int = np.array(np.random.choice(range(N),N,replace=False))
        index = []
        step = N//n_folds
        for i in np.arange(0,n_folds):
            j = np.arange(i*step,(i+1)*step)
            index.append(number_int[j])
            np.delete(number_int,j)
        return np.array(index)
                
```

```python
MyReg = Regression()
X_mean, X_std = MyReg.scalar(X)
X_train_sca = (X_train- X_mean)/X_std
```

```python
beta_hat_LS = MyReg.OLS(X_train_sca,y_train)
X_train_arg = MyReg.X_arg(X_train_sca)
N, p = X_train_arg.shape
sigma = 1/(N-(p-1)-1) * sum((y_train-X_train_arg@beta_hat_LS)**2)
v = np.diag(np.linalg.inv(X_train_arg.transpose()@X_train_arg))
z_score = beta_hat_LS/(np.sqrt(sigma)*np.sqrt(v))
std = np.sqrt(sigma)*np.sqrt(v)
```

### Table 3.2 

```python
table = zip(["Intercept"]+feature, beta_hat_LS, std, z_score)
print('      Term   Coefficient   Std. Error   Z Score')
print('-----------------------------------------------')
for term, coef, stde, zs in table:
    print(f'{term:>10}{coef:>14.2f}{stde:>13.2f}{zs:>10.2f}')
```
